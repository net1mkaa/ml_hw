```python
import random
import math

def objective_function(vars_list):
    x, y, z, w = vars_list
    return abs(50 - (2*x + 3*y + 5*z + 7*w))

def mutate(current_vars):
    mutated = list(current_vars)
    idx = random.randrange(4)
    mutated[idx] = max(1, mutated[idx] + random.choice([-3, -2, -1, 1, 2, 3]))
    return mutated

def optimize():
    random.seed(42)
    state = [random.randint(1, 10) for _ in range(4)]
    error = objective_function(state)
    
    temp = 1000.0
    alpha = 0.99
    steps = 0
    
    while temp > 0.001 and error > 0:
        new_state = mutate(state)
        new_error = objective_function(new_state)
        
        if new_error < error or random.random() < math.exp((error - new_error) / temp):
            state = new_state
            error = new_error
            
        temp *= alpha
        steps += 1
        
    return state, error, steps

final_vars, final_err, total_steps = optimize()
print(f"Result: x={final_vars[0]}, y={final_vars[1]}, z={final_vars[2]}, w={final_vars[3]}")
print(f"Final error: {final_err}")
print(f"Found in {total_steps} iterations.")
```
